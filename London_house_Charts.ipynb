```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path = r'C:\Users\SDCABASA\Desktop\My Data Analysis Projects\London_Housing Analysis\london_house_prices.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

### Average Price and Price Distribution

```python
# Calculate average price
average_price = df['price_pounds'].mean()

# Price distribution histogram
plt.figure(figsize=(10, 6))
plt.hist(df['price_pounds'], bins=20, color='skyblue', edgecolor='black')
plt.title('Price Distribution')
plt.xlabel('Price (Pounds)')
plt.ylabel('Frequency')
plt.axvline(average_price, color='red', linestyle='dashed', linewidth=2, label=f'Avg Price: £{average_price:.2f}')
plt.legend()
plt.show()
```

### Correlation between Bedrooms and Property Prices

```python
# Scatter plot for bedrooms vs. property prices
plt.figure(figsize=(10, 6))
plt.scatter(df['bedrooms'], df['price_pounds'], alpha=0.5)
plt.title('Correlation between Bedrooms and Property Prices')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price (Pounds)')
plt.grid(True)
plt.show()
```

### Price Comparison for Freehold vs. Leasehold

```python
# Calculate average prices for freehold and leasehold properties
avg_price_freehold = df[df['tenure'] == 'freehold']['price_pounds'].mean()
avg_price_leasehold = df[df['tenure'] == 'leasehold']['price_pounds'].mean()

# Bar chart to compare prices
plt.figure(figsize=(8, 6))
plt.bar(['Freehold', 'Leasehold'], [avg_price_freehold, avg_price_leasehold], color=['skyblue', 'lightcoral'])
plt.title('Average Property Prices by Tenure')
plt.ylabel('Average Price (Pounds)')
plt.show()
```

### Price Comparison for Properties with Gardens

```python
# Calculate average prices for properties with and without gardens
avg_price_with_garden = df[df['garden'] == 1]['price_pounds'].mean()
avg_price_without_garden = df[df['garden'] == 0]['price_pounds'].mean()

# Bar chart to compare prices
plt.figure(figsize=(8, 6))
plt.bar(['With Garden', 'Without Garden'], [avg_price_with_garden, avg_price_without_garden], color=['skyblue', 'lightcoral'])
plt.title('Average Property Prices with and without Gardens')
plt.ylabel('Average Price (Pounds)')
plt.show()
```
