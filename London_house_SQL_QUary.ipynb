### 1. What is the average price of properties in the dataset, and what is the price distribution?

```python
%%sql
SELECT AVG(price_pounds) AS average_price
FROM london_house_prices;

-- Price distribution
SELECT price_pounds, COUNT(*) AS frequency
FROM london_house_prices
GROUP BY price_pounds
ORDER BY price_pounds;
```

### 2. Is there a correlation between the number of bedrooms and property prices? How does the number of bathrooms affect property prices?


```python
%%sql
-- Correlation between bedrooms and prices
SELECT 
    (COUNT(*) * SUM(bedrooms * price_pounds) - SUM(bedrooms) * SUM(price_pounds)) / 
    (SQRT((COUNT(*) * SUM(bedrooms * bedrooms) - POW(SUM(bedrooms), 2)) * 
          (COUNT(*) * SUM(price_pounds * price_pounds) - POW(SUM(price_pounds), 2)))) 
    AS bedroom_price_correlation
FROM london_house_prices;
```

### 3. Do freehold properties tend to be more expensive than leasehold properties on average?


```python
%%sql
SELECT tenure, AVG(price_pounds) AS avg_price
FROM london_house_prices
GROUP BY tenure;
```

### 4. Are properties with gardens generally priced higher than those without gardens?

```python
%%sql
SELECT garden, AVG(price_pounds) AS avg_price
FROM london_house_prices
GROUP BY garden;
```

### 5. What is the relationship between the distance to the nearest station and property prices? Is proximity to public transportation a significant factor in property pricing?

```python
%%sql
SELECT nearest_station_miles, AVG(price_pounds) AS avg_price
FROM london_house_prices
GROUP BY nearest_station_miles
ORDER BY nearest_station_miles;
```

### 6. Can you identify any geographic trends in property prices based on the outer postcodes? Are certain areas of London more expensive than others?.

```python
%%sql 
SELECT postcode_outer, AVG(price_pounds) AS avg_price
FROM london_house_prices
GROUP BY postcode_outer
ORDER BY avg_price DESC;
```

### 7. What are the most common property types (e.g., apartments, houses) in the dataset, and how do their attributes and prices differ?

```python
%%sql
SELECT street, COUNT(*) AS property_count, AVG(size_sqft) AS avg_size, AVG(price_pounds) AS avg_price
FROM london_house_prices
GROUP BY street
ORDER BY property_count DESC;
```

### 8. Is there a correlation between property size (size_sqft) and price? Are larger properties consistently more expensive, or are there exceptions?

```python
%%sql
SELECT
    (COUNT(*) * SUM(size_sqft * price_pounds) - SUM(size_sqft) * SUM(price_pounds)) /
    (SQRT((COUNT(*) * SUM(size_sqft * size_sqft) - POW(SUM(size_sqft), 2)) *
          (COUNT(*) * SUM(price_pounds * price_pounds) - POW(SUM(price_pounds), 2))))
    AS size_price_correlation
FROM london_house_prices;
```
